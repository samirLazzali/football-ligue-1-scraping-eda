```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn
from scipy.stats import poisson,skellam
import os

%matplotlib inline
plt.style.use('seaborn-whitegrid') # theme
plt.rcParams['figure.figsize'] = (12,8) # taille des figures
ext = pd.read_csv("/home/sam/Documents/cours/python/football-ligue-1-scraping-eda-master/exctract.csv")
ext = ext.drop(['bmkr_loss','outcome_effectif', 'outcome_odds', 'prediction_odds'], axis=1)


ext.head(10)
```

```python
#Cration de outcome_effectif : resultat du match
def func(row):
    outcome_effectif = 'HOME'
    if(row.score_b > row.score_a):
        outcome_effectif = 'AWAY'
    elif(row.score_b == row.score_a) :
        outcome_effectif = 'DRAW'
    return outcome_effectif

ext['outcome_effectif'] = ext.apply(func, axis=1)
ext
```

```python
# min_odds : qui le bookmaker voyait gagnant (plus petite des cotes)
def func(row):
    # on prend l'index de la cote la plus petite : 
    min_index = [row.odds_a, row.odds_b, row.odds_draw].index(min([row.odds_a, row.odds_b, row.odds_draw]))
    # on la traduit en classe
    return ['HOME', 'AWAY', 'DRAW'][min_index]
    
    #prediction_odds = 1 if outcome_odds == outcome_effectif else 0 

ext['min_odds'] = ext.apply(func, axis=1)
ext
```

```python
# max_odds : qui le bookmaker voyait perdant (plus grande des cotes)
def func(row):
    # on prend l'index de la cote la plus petite : 
    max_index = [row.odds_a, row.odds_b, row.odds_draw].index(max([row.odds_a, row.odds_b, row.odds_draw]))
    # on la traduit en classe
    return ['HOME', 'AWAY', 'DRAW'][max_index]
ext['max_odds'] = ext.apply(func, axis=1)
#ext.head(5)
```

```python
# good_pred le bookmaker a prÃ©dit le resultat effectif
ext['good_pred'] =  pd.Series(ext['min_odds'] == ext['outcome_effectif'],dtype='int64')
# la plus grande cote l'emporte
ext['worst_pred'] =  pd.Series(ext['max_odds'] == ext['outcome_effectif'],dtype='int64')
ext['draw_pred'] = pd.Series((ext['good_pred']==0) & (ext['worst_pred'] == 0), dtype='int64')
ext[ext.draw_pred == 1].head(5)
```
